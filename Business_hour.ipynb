```python
business_hours_list1 = []
business_hours_list_1_start_time = []
business_hours_list_1_end_time = []

business_hours_list2 = []
business_hours_list_2_start_time = []
business_hours_list_2_end_time = []

business_hours_list3 = []
business_hours_list_3_start_time = []
business_hours_list_3_end_time = []

business_hours_list4 = []
business_hours_list_4_start_time = []
business_hours_list_4_end_time = []

business_hours_list5 = []
business_hours_list_5_start_time = []
business_hours_list_5_end_time = []

business_hours_list6 = []
business_hours_list_6_start_time = []
business_hours_list_6_end_time = []

business_hours_list7 = []
business_hours_list_7_start_time = []
business_hours_list_7_end_time = []

def convert_business_hours(business_hours):
    start_time, end_time = business_hours.split('-')#split and get start time
    start_hour, start_minute = map(int, start_time.split(':'))

    if 0<=start_hour<6: #set the start time class
        start_type = 'midnight'
    elif 6<=start_hour<12:
        start_type = 'morning'
    elif 12<=start_hour<18:
        start_type = 'afternoon'
    elif 18<=start_hour<24:
        start_type = 'night'

    end_hour, end_minute = end_time.split(':')#split and get end time
    end_hour= int(end_hour)
    end_minute = int(end_minute)

    if 0<=end_hour<6: #set the end time class
        end_type = 'midnight'
        # business_hours_list_1_end_time.append('midnight')
    elif 6<=end_hour<12:
        end_type = 'morning'
        # business_hours_list_1_end_time.append('morning')
    elif 12<=end_hour<18:
        end_type = 'afternoon'
        # business_hours_list_1_end_time.append('afternoon')
    elif 18<=end_hour<24:
        end_type = 'night'
        # business_hours_list_1_end_time.append('night')

    start_total_minutes = start_hour * 60 + start_minute #Unified units for minutes and hours
    end_total_minutes = end_hour * 60 + end_minute

    business_duration_minutes = end_total_minutes - start_total_minutes #After subtracting to obtain the difference, convert it into hourly units
    business_duration_hours = business_duration_minutes / 60

    if business_duration_hours <= 0:
        business_duration_hours = business_duration_hours + 24 #When time is negative

    return business_duration_hours , start_type , end_type

for value in data_time['Monday']:
    if value=='nan':
        business_hours_list1.append('none')
        business_hours_list_1_start_time.append('none')
        business_hours_list_1_end_time.append('none')
    else :
        working_hours1, start_type1 , end_type1 = convert_business_hours(value)
        business_hours_list1.append(working_hours1)
        business_hours_list_1_start_time.append(start_type1)
        business_hours_list_1_end_time.append(end_type1)

for value in data_time['Tuesday']:
    if value=='nan':
        business_hours_list2.append('none')
        business_hours_list_2_start_time.append('none')
        business_hours_list_2_end_time.append('none')
    else :
        working_hours2, start_type2, end_type2 = convert_business_hours(value)
        business_hours_list2.append(working_hours2)
        business_hours_list_2_start_time.append(start_type2)
        business_hours_list_2_end_time.append(end_type2)

for value in data_time['Wednesday']:
    if value=='nan':
        business_hours_list3.append('none')
        business_hours_list_3_start_time.append('none')
        business_hours_list_3_end_time.append('none')
    else :
        working_hours3, start_type3, end_type3 = convert_business_hours(value)
        business_hours_list3.append(working_hours3)
        business_hours_list_3_start_time.append(start_type3)
        business_hours_list_3_end_time.append(end_type3)

for value in data_time['Thursday']:
    if value=='nan':
        business_hours_list4.append('none')
        business_hours_list_4_start_time.append('none')
        business_hours_list_4_end_time.append('none')
    else :
        working_hours4, start_type4, end_type4 = convert_business_hours(value)
        business_hours_list4.append(working_hours4)
        business_hours_list_4_start_time.append(start_type4)
        business_hours_list_4_end_time.append(end_type4)

for value in data_time['Friday']:
    if value=='nan':
        business_hours_list5.append('none')
        business_hours_list_5_start_time.append('none')
        business_hours_list_5_end_time.append('none')
    else :
        working_hours5, start_type5, end_type5 = convert_business_hours(value)
        business_hours_list5.append(working_hours5)
        business_hours_list_5_start_time.append(start_type5)
        business_hours_list_5_end_time.append(end_type5)

for value in data_time['Saturday']:
    if value=='nan':
        business_hours_list6.append('none')
        business_hours_list_6_start_time.append('none')
        business_hours_list_6_end_time.append('none')
    else :
        working_hours6, start_type6, end_type6 = convert_business_hours(value)
        business_hours_list6.append(working_hours6)
        business_hours_list_6_start_time.append(start_type6)
        business_hours_list_6_end_time.append(end_type6)

for value in data_time['Sunday']:
    if value=='nan':
        business_hours_list7.append('none')
        business_hours_list_7_start_time.append('none')
        business_hours_list_7_end_time.append('none')
    else :
        working_hours7, start_type7, end_type7 = convert_business_hours(value)
        business_hours_list7.append(working_hours7)
        business_hours_list_7_start_time.append(start_type7)
        business_hours_list_7_end_time.append(end_type7)

data_time['Monday_hours'] = business_hours_list1
data_time['Monday_start'] = business_hours_list_1_start_time
data_time['Monday_end'] = business_hours_list_1_end_time

data_time['Tuesday_hours'] = business_hours_list2
data_time['Tuesday_start'] = business_hours_list_2_start_time
data_time['Tuesday_end'] = business_hours_list_2_end_time

data_time['Wednesday_hours'] = business_hours_list3
data_time['Wednesday_start'] = business_hours_list_3_start_time
data_time['Wednesday_end'] = business_hours_list_3_end_time

data_time['Thursday_hours'] = business_hours_list4
data_time['Thursday_start'] = business_hours_list_4_start_time
data_time['Thursday_end'] = business_hours_list_4_end_time

data_time['Friday_hours'] = business_hours_list5
data_time['Friday_start'] = business_hours_list_5_start_time
data_time['Friday_end'] = business_hours_list_5_end_time

data_time['Saturday_hours'] = business_hours_list6
data_time['Saturday_start'] = business_hours_list_6_start_time
data_time['Saturday_end'] = business_hours_list_6_end_time

data_time['Sunday_hours'] = business_hours_list7
data_time['Sunday_start'] = business_hours_list_7_start_time
data_time['Sunday_end'] = business_hours_list_7_end_time
```
